## Tratando os dados

### Quebrando o variavel data

```python
dataset['day'] = pd.Series()
dataset['month'] = pd.Series()
dataset['year'] = pd.Series()
```

```python
from datetime import datetime
```

```python
for i in dataset.index:
    dataset['day'][i] = datetime.strptime(dataset.Date_of_Journey[i], '%d/%m/%Y').day
    dataset['month'][i] = datetime.strptime(dataset.Date_of_Journey[i], '%d/%m/%Y').month
    dataset['year'][i] = datetime.strptime(dataset.Date_of_Journey[i], '%d/%m/%Y').year
```

```python
dataset.head()
```

```python
dataset = dataset.drop("Date_of_Journey", axis=1)
```

### Quebrando a variável hora

```python
dataset["Dep_hour"] = pd.Series()
dataset["Dep_min"] = pd.Series()
dataset["Arrival_hour"] = pd.Series()
dataset["Arrival_min"] = pd.Series()
```

```python
for i in dataset.index:
    dataset["Dep_hour"][i] = dataset["Dep_Time"][i][0:2]
    dataset["Dep_min"][i] = dataset["Dep_Time"][i][3:]
    dataset["Arrival_hour"][i] = dataset["Arrival_Time"][i][0:2]
    dataset["Arrival_min"][i] = dataset["Arrival_Time"][i][3:5]
```

```python
dataset.head()
```

```python
dataset = dataset.drop(["Dep_Time", "Arrival_Time"], axis=1)
```

```python
dataset.head()
```

### Quebrando a coluna duration

```python
dataset['Duration_hour'] = pd.Series()
dataset['Duration_min'] = pd.Series()
```

```python
for i in dataset.index:
    dataset['Duration_hour'][i] = dataset['Duration'][i].split('h')[0]
    if len(dataset['Duration'][i].split('h')) > 1:
        dataset['Duration_min'][i] = dataset['Duration'][i].split('h')[1][0:3]
    if dataset['Duration_min'][i] == '':
        dataset['Duration_min'][i] = '00'
    
```

```python
dataset.head()
```

```python
dataset = dataset.drop('Duration', axis=1)
```

```python
dataset.Total_Stops.unique()
```

### Quebrando a coluna Total_Stops

```python
dataset["stops"] = pd.Series()
```

```python
dataset["Total_Stops"] = dataset["Total_Stops"].fillna('0 stops')
```

```python
dataset.Total_Stops.unique()
```

```python
for i in dataset.index:
    if dataset['Total_Stops'][i] == 'non-stop':
        dataset["stops"][i] = '0'
    elif dataset['Total_Stops'][i] != '0':
        dataset["stops"][i] = dataset['Total_Stops'][i][0:1]
        
```

```python
dataset['stops'] = dataset['stops'].fillna('0')
```

```python
dataset = dataset.drop('Total_Stops', axis=1)
```

```python
dataset.head()
```

## Aplicando Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
columns = dataset.dtypes.reset_index()
columns
```

```python
categ_cols = columns[columns[0]=='object']['index'].to_list()
categ_cols
```

```python
for i in categ_cols:
    dataset[str(i)+'_encoded'] = le.fit_transform(dataset[i])
    dataset = dataset.drop(i, axis=1)
    
```

```python
dataset.head()
```

## Quebrando o dataset para modelagem

```python
from sklearn.model_selection import train_test_split

x = dataset.drop('Price', axis=1)
y = dataset['Price']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

## Aplicando o modelo

```python
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import r2_score
import time
```

```python
modelo_gr = GradientBoostingRegressor(max_depth=10, max_leaf_nodes=20, random_state=42)
modelo_xb = XGBRegressor(max_depth=10, max_leaf_nodes=20, random_state=42)
modelo_lgbm = LGBMRegressor(max_depth=10, max_leaf_nodes=20, random_state=42)
```

#### Avaliando o modelo de GradientBoosting

```python
start = time.time()
modelo_gr.fit(x_train, y_train)
y_pred_train = modelo_gr.predict(x_train)
y_pred_test = modelo_gr.predict(x_test)
end = time.time()
```

```python
print("Modelo de treino:", r2_score(y_train, y_pred_train))
print("Modelo de treino:", r2_score(y_test, y_pred_test))
print("Tempo:", str(end-start))
```

#### Avaliando o modelo XGBoost

```python
start = time.time()
modelo_xb.fit(x_train, y_train)
y_pred_train = modelo_xb.predict(x_train)
y_pred_test = modelo_xb.predict(x_test)
end = time.time()
```

```python
print("Modelo de treino:", r2_score(y_train, y_pred_train))
print("Modelo de treino:", r2_score(y_test, y_pred_test))
print("Tempo:", str(end-start))
```

#### Avaliando o modelo LGBM

```python
start = time.time()
modelo_lgbm.fit(x_train, y_train)
y_pred_train = modelo_lgbm.predict(x_train)
y_pred_test = modelo_lgbm.predict(x_test)
end = time.time()
```

```python
print("Modelo de treino:", r2_score(y_train, y_pred_train))
print("Modelo de treino:", r2_score(y_test, y_pred_test))
print("Tempo:", str(end-start))
```
