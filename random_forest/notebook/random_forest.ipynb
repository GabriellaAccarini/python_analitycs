## Tratamento dos dados

### Quebrando a variável de data

```python
# Criando as colunas 'day', 'month', 'year'

dataset['day'] = pd.Series()
dataset['month'] = pd.Series()
dataset['year'] = pd.Series()
```

```python
from datetime import datetime

for i in dataset.index:
    dataset['day'][i] = datetime.strptime(dataset.Date_of_Journey[i], '%d/%m/%Y').day
    dataset['month'][i] = datetime.strptime(dataset.Date_of_Journey[i], '%d/%m/%Y').month
    dataset['year'][i] = datetime.strptime(dataset.Date_of_Journey[i], '%d/%m/%Y').year
```

```python
dataset.head(5)
```

```python
dataset = dataset.drop("Date_of_Journey", axis=1)
```

### Quebrando a variável hora

```python
dataset["Dep_hour"] = pd.Series()
dataset["Dep_min"] = pd.Series()
dataset["Arrival_hour"] = pd.Series()
dataset["Arrival_min"] = pd.Series()
```

```python
for i in dataset.index:
    dataset["Dep_hour"][i] = dataset.Dep_Time[i][0:2]
    dataset["Dep_min"][i] = dataset.Dep_Time[i][3:6]
    dataset["Arrival_hour"][i] = dataset.Arrival_Time[i][0:2]
    dataset["Arrival_min"][i] = dataset.Arrival_Time[i][3:6]
```

```python
dataset = dataset.drop(["Dep_Time", "Arrival_Time"], axis=1)
```

```python
dataset.head(5)
```

### Tratando a quantidade de paradas

```python
dataset.Total_Stops.unique()
```

```python
dataset.Total_Stops = dataset.Total_Stops.fillna('non-stop')
dataset['stops'] = pd.Series()

for i in dataset.index:
    if dataset.Total_Stops[i]== 'non-stop':
        dataset.stops[i]=0
    else:
        dataset.stops[i]=dataset.Total_Stops[i][0:2]
        
        
```

```python
dataset.sample(5)
```

```python
dataset=dataset.drop("Total_Stops", axis=1)
```

### Aplicando Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

colunas = dataset.dtypes.reset_index()
colunas
```

```python
categ_cols = colunas[colunas[0]=='object']['index'].to_list()
categ_cols
```

```python
for i in categ_cols:
    dataset[i] = dataset[i].astype(str)
    dataset[str(i)+'_encoded'] = le.fit_transform(dataset[i])
    dataset = dataset.drop(i, axis=1)
```

```python
dataset.head()
```

## Quebrando o dataset para modelagem

```python
from sklearn.model_selection import train_test_split

x = dataset.drop('Price', axis=1)
y = dataset['Price']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

## Carregando o modelo

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
modelo = RandomForestRegressor(max_depth=10, max_leaf_nodes=20, random_state=42)
```

```python
modelo = modelo.fit(x_train, y_train)
y_pred_train = modelo.predict(x_train)
y_pred_test = modelo.predict(x_test)
```

```python
print('Resultado da base de treino é', r2_score(y_train, y_pred_train))
print('Resultado da base de teste é', r2_score(y_test, y_pred_test))
```

```python
### Fazendo uma tentativa de variar os parâmetros da árvore
```

```python
depth = [10,50,100,500]
leaf = [20,50,100,500]

for i,j in zip(depth,leaf):
    modelo = RandomForestRegressor(max_depth=i, max_leaf_nodes=j, random_state=42)
    modelo = modelo.fit(x_train, y_train)
    y_pred_train = modelo.predict(x_train)
    y_pred_test = modelo.predict(x_test)
    print('Resultado da base de treino é', r2_score(y_train, y_pred_train))
    print('Resultado da base de teste é', r2_score(y_test, y_pred_test))
    print('\n')
```

```python
### Conforme aumentado a profundidade da árvore, melhor a performance e não tivemos overfitting
```
